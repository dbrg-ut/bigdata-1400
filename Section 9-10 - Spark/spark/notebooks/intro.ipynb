```python
import pyspark

from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.master('spark://DESKTOP-HSK5ETQ.localdomain:7077').appName('Pi-Estimation').getOrCreate()
```

```python
import random

NUM_SAMPLES=10000

def inside(p):
    x,y = random.random(), random.random()
    return x*x + y*y < 1

data = spark.sparkContext.parallelize(range(0,NUM_SAMPLES)).filter(inside).collect()
print(data)
print('Pi is roughly {}'.format(4.0 * len(data) / NUM_SAMPLES))
```

```python
data = spark.sparkContext.parallelize(range(0,NUM_SAMPLES),10).filter(inside).collect()
print(data)
print('Pi is roughly {}'.format(4.0 * len(data) / NUM_SAMPLES))
```
